```python
import numpy as np
from sklearn.metrics import accuracy_score

class PerceptronModel:

    def __init__(self, lr=0.1, epochs=1000):
        self.lr = lr
        self.epochs = epochs
        self.w = None
        self.b = 0

    def train(self, X, Y):

        X = np.array(X)
        Y = np.array(Y)

        n_features = X.shape[1]
        self.w = np.zeros(n_features)

        for _ in range(self.epochs):

            for idx in range(len(X)):

                linear_output = np.dot(X[idx], self.w) + self.b
                prediction = self.activation(linear_output)

                diff = Y[idx] - prediction

                self.w += self.lr * diff * X[idx]
                self.b += self.lr * diff


    def predict(self, X):

        X = np.array(X)
        outputs = []

        for sample in X:
            value = np.dot(sample, self.w) + self.b
            outputs.append(1 if value >= 0 else 0)

        return np.array(outputs)


    def activation(self, x):
        return 1 if x >= 0 else 0


# Inputs for logic gates
X = [
    [0,0],
    [0,1],
    [1,0],
    [1,1]
]

logic_gates = {
    "AND":[0,0,0,1],
    "OR":[0,1,1,1],
    "NAND":[1,1,1,0],
    "NOR":[1,0,0,0]
}

for gate_name, target in logic_gates.items():

    print("\nGate:", gate_name)

    model = PerceptronModel()
    model.train(X, target)

    preds = model.predict(X)

    print("Predictions:", preds)
    print("Accuracy:", accuracy_score(target, preds))
```
